## W7A16
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
lst = list(map(int, input().split()))
dem = {}
for x in lst:
    dem[x] = dem.get(x, 0) + 1
result = sorted(x for x in dem if dem[x] == x)
print(result)
```

## W7A17
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore.
Nếu không có phần tử nào thỏa mãn thì không in gì

```python
lst = list(map(int, input().split()))
candidate, count = None, 0
for x in lst:
    if count == 0:
        candidate = x
    count += 1 if x == candidate else -1
if lst.count(candidate) > len(lst) // 2:
    print(candidate)
else:
    print("None")
```

## W7A19
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không
dùng bất cứ thuật toán sắp xếp nào

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))
giam = len(a) > 1 and a[0] > a[1]
if giam:
    a, b = a[::-1], b[::-1]
result = []
i = j = 0
while i < len(a) and j < len(b):
    if a[i] <= b[j]:
        result.append(a[i]); i += 1
    else:
        result.append(b[j]); j += 1
result += a[i:] + b[j:]
if giam:
    result = result[::-1]
print(result)
```

## W7A22
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất.
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
n = int(input())
lst = [input() for _ in range(n)]
ten = [x.split()[-1] for x in lst]
dem = {}
for t in ten:
    dem[t] = dem.get(t, 0) + 1
max_count = max(dem.values())
print(sorted(t for t in dem if dem[t] == max_count)[0])
```

## W7A23
Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài >=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái

```python
s = input()
words = s.split()
count = 0
for w in words:
    if len(w) >= 2 and w[0].isupper() and w[1:].islower() and w.isalpha():
        count += 1
print(count)
```

## W7A26
Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các số đó.

```python
s = input()
nums = []
i = 0
while i < len(s):
    if s[i].isdigit():
        j = i
        while j < len(s) and s[j].isdigit():
            j += 1
        nums.append(int(s[i:j]))
        i = j
    else:
        i += 1
if nums:
    print(f"Tong: {sum(nums)}, Trung binh: {sum(nums)/len(nums):.2f}")
```

## W7A27
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ
hay không

```python
s = input()
pairs = {')':'(', '}':'{', ']':'['}
stack = []
hop_le = True
for c in s:
    if c in '({[':
        stack.append(c)
    elif c in ')}]':
        if not stack or stack[-1] != pairs[c]:
            hop_le = False
            break
        stack.pop()
print(hop_le and len(stack) == 0)
```

## W7A28
Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
lst = list(map(int, input().split()))
N = int(input())
best, cur = [], []
for x in lst:
    if x > N:
        cur.append(x)
        if len(cur) > len(best):
            best = cur[:]
    else:
        cur = []
print(best)
```

## W7A29
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
lst = list(map(int, input().split()))
best, cur = [lst[0]], [lst[0]]
for i in range(1, len(lst)):
    if lst[i] > lst[i-1]:
        cur.append(lst[i])
        if len(cur) > len(best):
            best = cur[:]
    else:
        cur = [lst[i]]
print(best)
```

## W7A30
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập
vào từ bàn phím 

```python
lst = list(map(int, input().split()))
n = len(lst)
if n == 0:
    print([])
else:
    dp = [1] * n
    prev = [-1] * n
    for i in range(1, n):
        for j in range(i):
            if lst[j] < lst[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j
    idx = dp.index(max(dp))
    result = []
    while idx != -1:
        result.append(lst[idx])
        idx = prev[idx]
    print(result[::-1])
```
