## W8A16

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

shape_type = input()
if shape_type == "RECTANGLE":
    w, h = map(float, input().split())
    print(f"{Rectangle(w, h).area():.2f}")
else:
    r = float(input())
    print(f"{Circle(r).area():.2f}")
```

## W8A17

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        return f"{self.brand}-{self.speed}"

class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats
    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"

obj_type = input()
data = input().split()
if obj_type == "Vehicle":
    print(Vehicle(data[0], data[1]).info())
else:
    print(Car(data[0], data[1], data[2]).info())
```

## W8A18

```python
class Polygon:
    def perimeter(self):
        return 0

class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c

a, b, c = map(int, input().split())
print(Triangle(a, b, c).perimeter())
```

## W8A19

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def describe(self):
        return f"{self.name} {self.age}"

class Student(Person):
    def __init__(self, name, age, school):
        super().__init__(name, age)
        self.school = school
    def describe(self):
        return f"{self.name} {self.age} {self.school}"

obj_type = input()
data = input().split()
if obj_type == "Person":
    print(Person(data[0], data[1]).describe())
else:
    print(Student(data[0], data[1], data[2]).describe())
```

## W8A20

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance += self.balance * self.rate

balance, rate = map(float, input().split())
acc = SavingAccount(balance, rate)
acc.add_interest()
print(f"{acc.get_balance():.2f}")
```

## W8A22

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a * self.d - self.b * self.c

a, b, c, d = map(int, input().split())
m = Matrix2x2(a, b, c, d)
print(m.trace(), m.determinant())
```

## W8A25

```python
class PolynomialSum:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def sum_str(self):
        k = self.a + self.c
        m = self.b + self.d
        return f"{k}x + {m}"

a, b, c, d = map(int, input().split())
print(PolynomialSum(a, b, c, d).sum_str())
```

## W8A27

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def final_price(self):
        return self.price

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent
    def final_price(self):
        return self.price * (100 - self.percent) / 100

data = input().split()
name = data[0]
price = float(data[1])
percent = float(data[2])
print(f"{DiscountProduct(name, price, percent).final_price():.2f}")
```

## W8A28

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count

n = int(input())
ops = input().split()
c = Counter()
for op in ops:
    if op == "INC":
        c.inc()
    else:
        c.dec()
print(c.value())
```

## W8A29

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s

h, m, s = map(int, input().split())
print(Time(h, m, s).to_seconds())
```

## W8A30

```python
class Stack:
    def __init__(self):
        self.data = []
    def push(self, x):
        self.data.append(x)
    def pop(self):
        self.data.pop()
    def top(self):
        return self.data[-1]

n = int(input())
s = Stack()
for _ in range(n):
    line = input().split()
    if line[0] == "PUSH":
        s.push(int(line[1]))
    elif line[0] == "POP":
        s.pop()
    else:
        print(s.top())
```
