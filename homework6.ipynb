W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo 

```python
a = int(input())
c = [list(map(int, input().split())) for _ in range(a)]
cheo_chinh = [c[i][i] for i in range(a)]
cheo_phu = [c[i][a - 1 - i] for i in range(a)]
print("Đường chéo chính:", *cheo_chinh)
print("Đường chéo phụ:", *cheo_phu)
```

W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
def timk(a,k):
    for i in range(len(a)):
        if a[i]==k:
            return i
    return -1
a=list(map(int, input().split()))
b=int(input())
timk(a,b)
```

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
a=input().split()
b=len(a)
c={}
for i in range(b):
    if a[i] not in c:
        c[a[i]]=i
    else:
        c[a[i]].append(i)
d={key : tuple(value) for key, value in c.items()}
print(d)
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 số đầu,cuối.

```python
a = list(map(int, input().split()))
b = 0
for i in range(1, len(a) - 1):
    if a[i] > a[i-1] and a[i] > a[i+1]:
        b += a[i]
print(b)
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
a = list(map(int, input().split()))
b = sorted(list(set(a)))
c = b[::-1]
if a == b and len(set(a)) == len(a):
    print("Tăng thực sự")
elif a == c and len(set(a)) == len(a):
    print("Giảm thực sự")
else:
    print("Vô danh")
```
