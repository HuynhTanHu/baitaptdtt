## W9A16

```python
def generate(n, m, current, zeros, ones, results):
    if zeros == n and ones == m:
        results.append(current)
        return
    if zeros < n:
        generate(n, m, current + "0", zeros + 1, ones, results)
    if ones < m:
        generate(n, m, current + "1", zeros, ones + 1, results)

n, m = map(int, input().split())
results = []
generate(n, m, "", 0, 0, results)
for r in results:
    print(r)
```

## W9A17

```python
def longest_increasing_subsequence(arr):
    n = len(arr)
    dp = [1] * n
    prev = [-1] * n
    best = 0
    for i in range(1, n):
        for j in range(i):
            if arr[j] < arr[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j
        if dp[i] > dp[best]:
            best = i
    seq = []
    idx = best
    while idx != -1:
        seq.append(arr[idx])
        idx = prev[idx]
    return seq[::-1]

arr = list(map(int, input().split()))
print(*longest_increasing_subsequence(arr))
```

## W9A18

```python
def count_pairs(arr, k):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == k:
                count += 1
    return count

n = int(input())
arr = list(map(int, input().split()))
k = int(input())
print(count_pairs(arr, k))
```

## W9A19

```python
def remove_duplicates(arr):
    seen = set()
    result = []
    for x in arr:
        if x not in seen:
            seen.add(x)
            result.append(x)
    return result

arr = list(map(int, input().split()))
print(*remove_duplicates(arr))
```

## W9A20

```python
def normalize(s):
    # Bỏ khoảng trắng đầu cuối thủ công
    start = 0
    end = len(s) - 1
    while start <= end and s[start] == ' ':
        start += 1
    while end >= start and s[end] == ' ':
        end -= 1
    s = s[start:end+1]

    # Tách từ, bỏ khoảng trắng thừa giữa
    words = []
    word = ""
    for ch in s:
        if ch == ' ':
            if word:
                words.append(word)
                word = ""
        else:
            word += ch
    if word:
        words.append(word)

    # Viết hoa chữ cái đầu mỗi từ thủ công
    result = []
    for w in words:
        if w:
            first = w[0]
            if 'a' <= first <= 'z':
                first = chr(ord(first) - 32)
            result.append(first + w[1:])

    return ' '.join(result)

s = input()
print(normalize(s))
```

## W9A21

```python
def partition(arr, left, right):
    pivot = arr[right]
    i = left - 1
    for j in range(left, right):
        if arr[j] >= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[right] = arr[right], arr[i+1]
    return i + 1

def kth_largest(arr, left, right, k):
    if left == right:
        return arr[left]
    pivot_idx = partition(arr, left, right)
    rank = pivot_idx - left + 1
    if rank == k:
        return arr[pivot_idx]
    elif k < rank:
        return kth_largest(arr, left, pivot_idx - 1, k)
    else:
        return kth_largest(arr, pivot_idx + 1, right, k - rank)

n = int(input())
arr = list(map(int, input().split()))
k = int(input())
print(kth_largest(arr, 0, n - 1, k))
```

## W9A22

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    count = {}
    for c in s1:
        count[c] = count.get(c, 0) + 1
    for c in s2:
        count[c] = count.get(c, 0) - 1
        if count[c] < 0:
            return False
    return True

s1 = input()
s2 = input()
print(is_anagram(s1, s2))
```

## W9A23

```python
def largest_k_digits(num_str, k):
    remove = len(num_str) - k
    stack = []
    for digit in num_str:
        while remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            remove -= 1
        stack.append(digit)
    return ''.join(stack[:k])

num_str = input()
k = int(input())
print(largest_k_digits(num_str, k))
```

## W9A24    

```python
def compress(s):
    result = ""
    i = 0
    while i < len(s):
        ch = s[i]
        count = 1
        while i + count < len(s) and s[i + count] == ch:
            count += 1
        result += ch + str(count)
        i += count
    return result

s = input()
print(compress(s))
```

## W9A25

```python
def is_strong_password(s):
    if len(s) < 8:
        return False
    has_lower = any('a' <= c <= 'z' for c in s)
    has_upper = any('A' <= c <= 'Z' for c in s)
    has_digit = any('0' <= c <= '9' for c in s)
    has_special = any(c in '!@#$' for c in s)
    return has_lower and has_upper and has_digit and has_special

s = input()
print(is_strong_password(s))
```
